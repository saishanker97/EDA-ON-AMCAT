### Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
top_ten_Specializations = df['Specialization'].value_counts().head(10).index

filtered_df = df[df['Specialization'].isin(top_ten_Specializations)]

grouped_data = pd.crosstab(filtered_df['Specialization'], filtered_df['Gender'])

grouped_data_percentage = grouped_data.div(grouped_data.sum(1), axis=0) * 100

ax = grouped_data_percentage.plot(kind='bar', stacked=False, figsize=(20, 6))

plt.title('Percentage of Persons by Specialization and Gender')
plt.xlabel('Specialization')
# plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='Gender')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.1f}%', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

- yes,there is a relationship between gender and specilization. Irrespective  of specialization male condidates are more than female candidates,

### Is there a relationship between gender and CollegeState? (i.e. Does the preference of CollegeState depend on the Gender?)

```python
top_ten_Specializations = df['CollegeState'].value_counts().head(10).index

filtered_df = df[df['CollegeState'].isin(top_ten_Specializations)]

grouped_data = pd.crosstab(filtered_df['CollegeState'], filtered_df['Gender'])

grouped_data_percentage = grouped_data.div(grouped_data.sum(1), axis=0) * 100

ax = grouped_data_percentage.plot(kind='bar', stacked=False, figsize=(20, 6))

plt.title('Percentage of Persons by CollegeState and Gender')
plt.xlabel('CollegeState')
# plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Gender')

for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:.1f}%', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```
